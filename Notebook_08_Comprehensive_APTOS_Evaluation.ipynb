## 2. Load Models & Test Set

```python
# ============================================================
# 2a. Model architectures (TWO different architectures)
# ============================================================

# ── Architecture 1: K-Fold model (custom 3-layer head) ──
class DRBinaryModel(nn.Module):
    """K-Fold model: backbone + custom 3-layer head, output [B]"""
    def __init__(self, model_name='efficientnet_b3', pretrained=False):
        super().__init__()
        self.backbone = timm.create_model(model_name, pretrained=pretrained, num_classes=0)
        in_features = self.backbone.num_features
        self.head = nn.Sequential(
            nn.Linear(in_features, 256),
            nn.LayerNorm(256),
            nn.GELU(),
            nn.Dropout(0.4),
            nn.Linear(256, 64),
            nn.LayerNorm(64),
            nn.GELU(),
            nn.Dropout(0.3),
            nn.Linear(64, 1),
        )

    def forward(self, x):
        features = self.backbone(x)
        return self.head(features).squeeze(-1)  # Output: [B]


# ── Architecture 2: 2-Phase model (simple head via model.classifier) ──
def load_twophase_model(ckpt_path, model_name='efficientnet_b3', device='cpu'):
    model = timm.create_model(model_name, pretrained=False)
    n_features = model.classifier.in_features
    model.classifier = nn.Sequential(
        nn.Dropout(p=0.4),
        nn.Linear(n_features, 1),
    )
    ckpt = torch.load(ckpt_path, map_location='cpu', weights_only=False)
    model.load_state_dict(ckpt['model_state_dict'])
    model = model.to(device)
    model.eval()
    return model, ckpt


class DRDataset(Dataset):
    def __init__(self, df, transform, root=Path(".")):
        self.df = df.reset_index(drop=True)
        self.transform = transform
        self.root = root

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = cv2.imread(str(self.root / row["preprocessed_path"]))
        if img is None:
            raise FileNotFoundError(f"Cannot read: {self.root / row['preprocessed_path']}")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = self.transform(image=img)["image"]
        label = torch.tensor(row["binary_label"], dtype=torch.float32)
        return img, label


# Transforms
val_transform = A.Compose([
    A.Resize(CONFIG["image_size"], CONFIG["image_size"]),
    A.Normalize(mean=CONFIG["mean"], std=CONFIG["std"]),
    ToTensorV2(),
])

tta_transform = A.Compose([
    A.Resize(CONFIG["image_size"], CONFIG["image_size"]),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.Normalize(mean=CONFIG["mean"], std=CONFIG["std"]),
    ToTensorV2(),
])

print("Model architectures & dataset class defined ✓")
```

```python
# ============================================================
# 2b. Load APTOS-only test set
# ============================================================
df_test = pd.read_csv(SPLITS_DIR / "test_split.csv")
print(f"Test set: {len(df_test)} images (APTOS 2019 only)")
print(f"\nBinary distribution:")
print(df_test["binary_label"].value_counts().sort_index().to_string())
print(f"\nDR grade distribution:")
print(df_test["dr_grade"].value_counts().sort_index().to_string())

missing = [p for p in df_test["preprocessed_path"] if not (ROOT / p).exists()]
assert len(missing) == 0, f"{len(missing)} files missing!"
print(f"\nAll {len(df_test)} files verified ✓")
```

```python
# ============================================================
# 2c. Load the 2-Phase single model
# ============================================================
assert TWOPHASE_CKPT.exists(), f"Missing: {TWOPHASE_CKPT}"
twophase_model, twophase_ckpt = load_twophase_model(
    TWOPHASE_CKPT, CONFIG["model_name"], DEVICE
)
print(f"2-Phase model loaded ✓")
print(f"  Checkpoint keys: {list(twophase_ckpt.keys())}")
if 'best_val_auc' in twophase_ckpt:
    print(f"  Best val AUC: {twophase_ckpt['best_val_auc']:.4f}")
tp_params = sum(p.numel() for p in twophase_model.parameters())
print(f"  Parameters: {tp_params:,}")
```

```python
# ============================================================
# 2d. Load all 5 K-Fold models
# ============================================================
fold_models = []
for fold_idx in range(1, CONFIG["n_folds"] + 1):
    path = KFOLD_DIR / f"best_{CONFIG['model_name']}_fold{fold_idx}.pth"
    assert path.exists(), f"Missing: {path}"
    ckpt = torch.load(path, map_location="cpu", weights_only=False)
    model = DRBinaryModel(model_name=CONFIG["model_name"], pretrained=False)
    model.load_state_dict(ckpt["state_dict"])
    model = model.to(DEVICE)
    model.eval()
    fold_models.append(model)
    print(f"  Fold {fold_idx}: loaded (val AUC = {ckpt.get('best_val_auc', 'N/A'):.4f})")

print(f"\n✓ All {CONFIG['n_folds']} fold models loaded")
print(f"  Total: 1 (2-phase) + {CONFIG['n_folds']} (K-Fold) = {CONFIG['n_folds']+1} models")
```

## 3. Test Set Inference — All Evaluation Levels

**Level A0:** 2-Phase single model (no TTA)  
**Level A1:** 2-Phase + 8× TTA  
**Level A2:** 2-Phase + TTA + Optimized Threshold  
**Level B1–B5:** Each K-Fold predicts individually (no TTA)  
**Level C:** Average of 5-fold probabilities (soft voting, no TTA)  
**Level D:** 5-fold + 8× TTA per fold  
**Level E:** Level D + Youden's J optimized threshold

```python
# ============================================================
# 3. Inference at all levels
# ============================================================
y_true = df_test["binary_label"].values.astype(int)

# ── Helper: basic inference for K-Fold model (output [B]) ──
@torch.no_grad()
def predict_kfold(model, df, transform):
    model.eval()
    ds = DRDataset(df, transform, ROOT)
    dl = DataLoader(ds, batch_size=CONFIG["batch_size"], shuffle=False,
                    num_workers=CONFIG["num_workers"], pin_memory=CONFIG["pin_memory"])
    all_logits = []
    for imgs, _ in dl:
        logits = model(imgs.to(DEVICE))
        all_logits.append(logits.cpu())
    return 1 / (1 + np.exp(-torch.cat(all_logits).numpy()))


# ── Helper: basic inference for 2-Phase model (output [B,1]) ──
@torch.no_grad()
def predict_twophase(model, df, transform):
    model.eval()
    ds = DRDataset(df, transform, ROOT)
    dl = DataLoader(ds, batch_size=CONFIG["batch_size"], shuffle=False,
                    num_workers=CONFIG["num_workers"], pin_memory=CONFIG["pin_memory"])
    all_logits = []
    for imgs, _ in dl:
        logits = model(imgs.to(DEVICE)).squeeze(-1)  # [B,1] → [B]
        all_logits.append(logits.cpu())
    return 1 / (1 + np.exp(-torch.cat(all_logits).numpy()))


# ── Helper: TTA for K-Fold model ──
@torch.no_grad()
def predict_tta_kfold(model, df, n_tta=8):
    model.eval()
    all_probs = []
    for i in range(n_tta):
        tfm = val_transform if i == 0 else tta_transform
        ds = DRDataset(df, tfm, ROOT)
        dl = DataLoader(ds, batch_size=CONFIG["batch_size"], shuffle=False,
                        num_workers=CONFIG["num_workers"], pin_memory=CONFIG["pin_memory"])
        logits = []
        for imgs, _ in dl:
            logits.append(model(imgs.to(DEVICE)).cpu())
        logits = torch.cat(logits).numpy()
        all_probs.append(1 / (1 + np.exp(-logits)))
    return np.mean(all_probs, axis=0)


# ── Helper: TTA for 2-Phase model (output [B,1] → squeeze) ──
@torch.no_grad()
def predict_tta_twophase(model, df, n_tta=8):
    model.eval()
    all_probs = []
    for i in range(n_tta):
        tfm = val_transform if i == 0 else tta_transform
        ds = DRDataset(df, tfm, ROOT)
        dl = DataLoader(ds, batch_size=CONFIG["batch_size"], shuffle=False,
                        num_workers=CONFIG["num_workers"], pin_memory=CONFIG["pin_memory"])
        logits = []
        for imgs, _ in dl:
            logits.append(model(imgs.to(DEVICE)).squeeze(-1).cpu())
        logits = torch.cat(logits).numpy()
        all_probs.append(1 / (1 + np.exp(-logits)))
    return np.mean(all_probs, axis=0)


# ══════════════════════════════════════════════════════════════
# LEVEL A0: 2-Phase single model (no TTA)
# ══════════════════════════════════════════════════════════════
print("Level A0: 2-Phase single model (no TTA)...")
t0 = time.time()
twophase_probs = predict_twophase(twophase_model, df_test, val_transform)
print(f"  Done: {time.time()-t0:.1f}s")

# ══════════════════════════════════════════════════════════════
# LEVEL A1: 2-Phase + TTA
# ══════════════════════════════════════════════════════════════
print("\nLevel A1: 2-Phase + 8x TTA...")
t0 = time.time()
twophase_probs_tta = predict_tta_twophase(twophase_model, df_test, CONFIG["tta_times"])
print(f"  Done: {time.time()-t0:.1f}s")

# ══════════════════════════════════════════════════════════════
# LEVEL A2: 2-Phase + TTA + OptThresh
# ══════════════════════════════════════════════════════════════
print("\nLevel A2: 2-Phase threshold optimization (Youden's J)...")
fpr_tp, tpr_tp, thresh_tp = roc_curve(y_true, twophase_probs_tta)
J_tp = tpr_tp - fpr_tp
opt_idx_tp = np.argmax(J_tp)
opt_threshold_tp = thresh_tp[opt_idx_tp]
print(f"  2-Phase optimal threshold: {opt_threshold_tp:.4f} (J = {J_tp[opt_idx_tp]:.4f})")

# ══════════════════════════════════════════════════════════════
# LEVEL B: Individual fold predictions (no TTA)
# ══════════════════════════════════════════════════════════════
print("\nLevel B: Individual fold inference (no TTA)...")
fold_probs_noTTA = {}
for i, mdl in enumerate(fold_models):
    t0 = time.time()
    fold_probs_noTTA[f"Fold {i+1}"] = predict_kfold(mdl, df_test, val_transform)
    print(f"  Fold {i+1}: {time.time()-t0:.1f}s")

# ══════════════════════════════════════════════════════════════
# LEVEL C: 5-Fold Ensemble (no TTA)
# ══════════════════════════════════════════════════════════════
print("\nLevel C: 5-Fold Ensemble (no TTA)...")
ensemble_noTTA = np.mean(list(fold_probs_noTTA.values()), axis=0)

# ══════════════════════════════════════════════════════════════
# LEVEL D: 5-Fold Ensemble + TTA
# ══════════════════════════════════════════════════════════════
print("\nLevel D: 5-Fold Ensemble + 8x TTA...")
fold_probs_TTA = {}
for i, mdl in enumerate(fold_models):
    t0 = time.time()
    fold_probs_TTA[f"Fold {i+1}"] = predict_tta_kfold(mdl, df_test, CONFIG["tta_times"])
    print(f"  Fold {i+1}: {time.time()-t0:.1f}s")

ensemble_TTA = np.mean(list(fold_probs_TTA.values()), axis=0)

# ══════════════════════════════════════════════════════════════
# LEVEL E: Ensemble + TTA + OptThresh
# ══════════════════════════════════════════════════════════════
print("\nLevel E: Ensemble threshold optimization (Youden's J)...")
fpr_ens, tpr_ens, thresh_ens = roc_curve(y_true, ensemble_TTA)
J_ens = tpr_ens - fpr_ens
opt_idx_ens = np.argmax(J_ens)
opt_threshold_ens = thresh_ens[opt_idx_ens]
print(f"  Ensemble optimal threshold: {opt_threshold_ens:.4f} (J = {J_ens[opt_idx_ens]:.4f})")

# ── Store all probabilities ──
all_predictions = {}
all_predictions["2-Phase (no TTA)"]       = {"probs": twophase_probs,     "threshold": 0.5}
all_predictions["2-Phase + TTA"]           = {"probs": twophase_probs_tta, "threshold": 0.5}
all_predictions["2-Phase + TTA + OptThr"]  = {"probs": twophase_probs_tta, "threshold": opt_threshold_tp}
for name, probs in fold_probs_noTTA.items():
    all_predictions[name] = {"probs": probs, "threshold": 0.5}
all_predictions["Ensemble (no TTA)"]       = {"probs": ensemble_noTTA,     "threshold": 0.5}
all_predictions["Ensemble + TTA"]          = {"probs": ensemble_TTA,       "threshold": 0.5}
all_predictions["Ensemble + TTA + OptThr"] = {"probs": ensemble_TTA,       "threshold": opt_threshold_ens}

print(f"\n✓ All inference complete. {len(all_predictions)} evaluation levels ready.")
```

## 4. Core Metrics + Bootstrap 95% Confidence Intervals

# Assistant
The selected code defines four lambda functions that create different evaluation metrics for a binary classification model. Here's a step-by-step explanation:

1. `_mk_acc = lambda yt, yp: accuracy_score(yt, (yp >= threshold).astype(int))` - Creates a function that calculates accuracy by comparing true labels (`yt`) with predicted probabilities (`yp`) that are converted to binary predictions using the threshold.

2. `_mk_sens = lambda yt, yp: recall_score(yt, (yp >= threshold).astype(int), zero_division=0)` - Creates a function that calculates sensitivity (recall for the positive class) with zero_division=0 to handle cases where there are no positive examples.

3. `_mk_spec = lambda yt, yp: recall_score(yt, (yp >= threshold).astype(int), pos_label=0, zero_division=0)` - Creates a function that calculates specificity (recall for the negative class) by setting pos_label=0.

4. `_mk_f1 = lambda yt, yp: f1_score(yt, (yp >= threshold).astype(int), zero_division=0)` - Creates a function that calculates the F1 score, which is the harmonic mean of precision and recall.

These functions are designed to be passed to a bootstrap confidence interval calculation function to estimate uncertainty in the model's performance metrics.

```python
# ============================================================
# 4. Comprehensive metrics for ALL evaluation levels
# ============================================================

def bootstrap_ci(y_true, y_score, metric_fn, n_boot=2000, ci=95, seed=42):
    rng = np.random.RandomState(seed)
    scores = []
    for _ in range(n_boot):
        idx = rng.choice(len(y_true), len(y_true), replace=True)
        if len(np.unique(y_true[idx])) < 2:
            continue
        try:
            scores.append(metric_fn(y_true[idx], y_score[idx]))
        except:
            continue
    alpha = (100 - ci) / 2
    return np.percentile(scores, alpha), np.percentile(scores, 100 - alpha)


def compute_full_metrics(y_true, probs, threshold=0.5, name="", do_bootstrap=False, n_boot=2000):
    preds = (probs >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, preds).ravel()

    m = {
        "Model": name,
        "Threshold": threshold,
        "AUC-ROC": roc_auc_score(y_true, probs),
        "Accuracy": accuracy_score(y_true, preds),
        "Balanced Acc": balanced_accuracy_score(y_true, preds),
        "Sensitivity": tp / (tp + fn) if (tp + fn) > 0 else 0,
        "Specificity": tn / (tn + fp) if (tn + fp) > 0 else 0,
        "Precision": tp / (tp + fp) if (tp + fp) > 0 else 0,
        "NPV": tn / (tn + fn) if (tn + fn) > 0 else 0,
        "F1-Score": f1_score(y_true, preds, zero_division=0),
        "Cohen Kappa": cohen_kappa_score(y_true, preds),
        "MCC": matthews_corrcoef(y_true, preds),
        "Brier Score": brier_score_loss(y_true, probs),
        "Log Loss": log_loss(y_true, probs),
        "TP": int(tp), "TN": int(tn), "FP": int(fp), "FN": int(fn),
    }
    prec_c, rec_c, _ = precision_recall_curve(y_true, probs)
    m["AUC-PR"] = auc(rec_c, prec_c)

    if do_bootstrap:
        m["AUC-ROC CI"] = bootstrap_ci(y_true, probs, roc_auc_score, n_boot)
        _mk_acc  = lambda yt, yp: accuracy_score(yt, (yp >= threshold).astype(int))
        _mk_sens = lambda yt, yp: recall_score(yt, (yp >= threshold).astype(int), zero_division=0)
        _mk_spec = lambda yt, yp: recall_score(yt, (yp >= threshold).astype(int), pos_label=0, zero_division=0)
        _mk_f1   = lambda yt, yp: f1_score(yt, (yp >= threshold).astype(int), zero_division=0)
        m["Accuracy CI"]    = bootstrap_ci(y_true, probs, _mk_acc, n_boot)
        m["Sensitivity CI"] = bootstrap_ci(y_true, probs, _mk_sens, n_boot)
        m["Specificity CI"] = bootstrap_ci(y_true, probs, _mk_spec, n_boot)
        m["F1 CI"]          = bootstrap_ci(y_true, probs, _mk_f1, n_boot)

    return m


N_BOOT = CONFIG["bootstrap_n"]
all_metrics = []
for name, pred_info in all_predictions.items():
    do_boot = ("2-Phase" in name and "no TTA" not in name) or ("Ensemble" in name)
    m = compute_full_metrics(y_true, pred_info["probs"], pred_info["threshold"],
                             name=name, do_bootstrap=do_boot, n_boot=N_BOOT)
    all_metrics.append(m)

comp_cols = ["Model", "Threshold", "AUC-ROC", "AUC-PR", "Accuracy", "Sensitivity",
             "Specificity", "Precision", "F1-Score", "Cohen Kappa", "MCC", "Brier Score"]
df_comp = pd.DataFrame(all_metrics)[comp_cols]

print("=" * 115)
print("  COMPREHENSIVE COMPARISON — ALL EVALUATION LEVELS ON APTOS TEST SET (531 images)")
print("=" * 115)
print(df_comp.to_string(index=False, float_format="{:.4f}".format))
print("=" * 115)

print("\n  Bootstrap 95% Confidence Intervals:")
for m in all_metrics:
    if "AUC-ROC CI" in m:
        print(f"\n  {m['Model']}:")
        for metric_name in ["AUC-ROC", "Accuracy", "Sensitivity", "Specificity", "F1"]:
            ci_key = f"{metric_name} CI"
            if ci_key in m:
                lo, hi = m[ci_key]
                val = m.get(metric_name, m.get(f"{metric_name}-Score", 0))
                print(f"    {metric_name:<20s}: {val:.4f} [{lo:.4f}, {hi:.4f}]")
```

## 8. Threshold Sensitivity Analysis

```python
# ============================================================
# 8. Threshold sweep — BOTH 2-Phase+TTA and Ensemble+TTA
# ============================================================
thresholds_sweep = np.linspace(0.01, 0.99, 200)

def sweep_thresholds(y_true, probs, thresholds):
    sweep = {"threshold": [], "sensitivity": [], "specificity": [],
             "precision": [], "npv": [], "f1": [], "accuracy": [], "youden_j": []}
    for t in thresholds:
        yp = (probs >= t).astype(int)
        tn_, fp_, fn_, tp_ = confusion_matrix(y_true, yp).ravel()
        sens_ = tp_ / (tp_ + fn_) if (tp_ + fn_) > 0 else 0
        spec_ = tn_ / (tn_ + fp_) if (tn_ + fp_) > 0 else 0
        prec_ = tp_ / (tp_ + fp_) if (tp_ + fp_) > 0 else 0
        npv_  = tn_ / (tn_ + fn_) if (tn_ + fn_) > 0 else 0
        f1_   = 2*prec_*sens_ / (prec_+sens_) if (prec_+sens_) > 0 else 0
        acc_  = (tp_ + tn_) / len(y_true)
        sweep["threshold"].append(t)
        sweep["sensitivity"].append(sens_)
        sweep["specificity"].append(spec_)
        sweep["precision"].append(prec_)
        sweep["npv"].append(npv_)
        sweep["f1"].append(f1_)
        sweep["accuracy"].append(acc_)
        sweep["youden_j"].append(sens_ + spec_ - 1)
    return pd.DataFrame(sweep)

df_sweep_tp  = sweep_thresholds(y_true, twophase_probs_tta, thresholds_sweep)
df_sweep_ens = sweep_thresholds(y_true, ensemble_TTA, thresholds_sweep)

fig, axes = plt.subplots(2, 2, figsize=(20, 14))
fig.suptitle("Threshold Sensitivity — 2-Phase+TTA vs Ensemble+TTA (APTOS Test Set)",
             fontsize=15, fontweight="bold", y=1.02)

for row_idx, (df_sw, opt_t, title_pfx) in enumerate([
    (df_sweep_tp, opt_threshold_tp, "2-Phase + TTA"),
    (df_sweep_ens, opt_threshold_ens, "Ensemble + TTA"),
]):
    ax = axes[row_idx, 0]
    ax.plot(df_sw["threshold"], df_sw["sensitivity"], lw=2, label="Sensitivity", color=PALETTE["dr"])
    ax.plot(df_sw["threshold"], df_sw["specificity"], lw=2, label="Specificity", color=PALETTE["no_dr"])
    ax.plot(df_sw["threshold"], df_sw["precision"], lw=2, label="Precision", color=PALETTE["blue"])
    ax.plot(df_sw["threshold"], df_sw["npv"], lw=2, label="NPV", color=PALETTE["purple"])
    ax.axvline(x=opt_t, color=PALETTE["orange"], ls="--", lw=2, label=f"Optimal t={opt_t:.3f}")
    ax.axvline(x=0.5, color=PALETTE["gray"], ls=":", lw=1.5, label="Default t=0.500")
    ax.set_xlabel("Threshold"); ax.set_ylabel("Metric Value")
    ax.set_title(f"{title_pfx} — Metrics vs Threshold", fontweight="bold")
    ax.legend(fontsize=8); ax.set_xlim([0, 1])

    ax = axes[row_idx, 1]
    ax.plot(df_sw["threshold"], df_sw["f1"], lw=2, label="F1-Score", color=PALETTE["blue"])
    ax.plot(df_sw["threshold"], df_sw["accuracy"], lw=2, label="Accuracy", color=PALETTE["no_dr"])
    ax.plot(df_sw["threshold"], df_sw["youden_j"], lw=2, label="Youden's J", color=PALETTE["orange"])
    ax.axvline(x=opt_t, color=PALETTE["orange"], ls="--", lw=2, label=f"Optimal t={opt_t:.3f}")
    ax.axvline(x=0.5, color=PALETTE["gray"], ls=":", lw=1.5, label="Default t=0.500")
    ax.set_xlabel("Threshold"); ax.set_ylabel("Score")
    ax.set_title(f"{title_pfx} — F1/Acc/J vs Threshold", fontweight="bold")
    ax.legend(fontsize=8); ax.set_xlim([0, 1])

plt.tight_layout()
plt.savefig(FIG_DIR / "threshold_analysis.png", dpi=300, bbox_inches="tight")
plt.show()
print(f"Saved → {FIG_DIR / 'threshold_analysis.png'}")
```

```python
# ============================================================
# 10. Subgroup analysis — best of each family
# ============================================================
grade_names = {0: "No DR", 1: "Mild", 2: "Moderate", 3: "Severe", 4: "Proliferative"}
grade_colors = ["#2ecc71", "#f1c40f", "#e67e22", "#e74c3c", "#8e44ad"]

# Best 2-Phase: +TTA+OptThresh
y_pred_tp_best = (twophase_probs_tta >= opt_threshold_tp).astype(int)
# Best Ensemble: +TTA+OptThresh
y_pred_ens_best = (ensemble_TTA >= opt_threshold_ens).astype(int)

df_analysis = df_test.copy()
df_analysis["prob_tp"]      = twophase_probs_tta
df_analysis["pred_tp"]      = y_pred_tp_best
df_analysis["correct_tp"]   = (df_analysis["binary_label"] == y_pred_tp_best).astype(int)
df_analysis["prob_ens"]     = ensemble_TTA
df_analysis["pred_ens"]     = y_pred_ens_best
df_analysis["correct_ens"]  = (df_analysis["binary_label"] == y_pred_ens_best).astype(int)

print("=" * 95)
print("  SUBGROUP ANALYSIS BY DR GRADE — 2-Phase+TTA+Opt vs Ensemble+TTA+Opt (APTOS Test)")
print("=" * 95)
print(f"\n  {'Grade':<16s} {'N':>5s} {'Bin':>5s} "
      f"{'2Ph P(DR)':>10s} {'2Ph Acc':>8s} {'2Ph Err':>8s} "
      f"{'Ens P(DR)':>10s} {'Ens Acc':>8s} {'Ens Err':>8s}")
print(f"  {'-'*82}")

subgroup_stats = []
for grade in sorted(df_analysis["dr_grade"].unique()):
    sub = df_analysis[df_analysis["dr_grade"] == grade]
    n = len(sub)
    binary = int(sub["binary_label"].iloc[0])
    tp_m, tp_a, tp_e = sub["prob_tp"].mean(), sub["correct_tp"].mean(), (sub["correct_tp"]==0).sum()
    en_m, en_a, en_e = sub["prob_ens"].mean(), sub["correct_ens"].mean(), (sub["correct_ens"]==0).sum()
    subgroup_stats.append({
        "grade": grade, "name": grade_names[grade], "n": n, "binary": binary,
        "tp_mean_prob": tp_m, "tp_acc": tp_a, "tp_errors": int(tp_e),
        "ens_mean_prob": en_m, "ens_acc": en_a, "ens_errors": int(en_e),
    })
    print(f"  {grade_names[grade]:<16s} {n:>5d} {binary:>5d} "
          f"{tp_m:>10.4f} {tp_a:>8.4f} {int(tp_e):>8d} "
          f"{en_m:>10.4f} {en_a:>8.4f} {int(en_e):>8d}")

# ── Visualization ──
fig, axes = plt.subplots(2, 3, figsize=(21, 12))
fig.suptitle("Subgroup Analysis — 2-Phase+TTA+Opt vs Ensemble+TTA+Opt",
             fontsize=15, fontweight="bold", y=1.02)

grade_list = sorted(df_analysis["dr_grade"].unique())
names = [grade_names[g] for g in grade_list]
x = np.arange(len(names))
width = 0.35

# Row 0: Prob box plots + accuracy + errors
ax = axes[0, 0]
data_tp = [df_analysis[df_analysis["dr_grade"]==g]["prob_tp"].values for g in grade_list]
bp = ax.boxplot(data_tp, labels=names, patch_artist=True, widths=0.6)
for patch, c in zip(bp["boxes"], grade_colors[:len(data_tp)]): patch.set_facecolor(c); patch.set_alpha(0.7)
ax.axhline(y=opt_threshold_tp, color=PALETTE["twophase"], ls="--", lw=1.5)
ax.set_ylabel("P(DR)"); ax.set_title("2-Phase+TTA+Opt: P(DR) by Grade", fontweight="bold")

ax = axes[0, 1]
data_ens = [df_analysis[df_analysis["dr_grade"]==g]["prob_ens"].values for g in grade_list]
bp = ax.boxplot(data_ens, labels=names, patch_artist=True, widths=0.6)
for patch, c in zip(bp["boxes"], grade_colors[:len(data_ens)]): patch.set_facecolor(c); patch.set_alpha(0.7)
ax.axhline(y=opt_threshold_ens, color=PALETTE["orange"], ls="--", lw=1.5)
ax.set_ylabel("P(DR)"); ax.set_title("Ensemble+TTA+Opt: P(DR) by Grade", fontweight="bold")

ax = axes[0, 2]
b1 = ax.bar(x - width/2, [s["tp_acc"] for s in subgroup_stats], width,
            color=PALETTE["twophase"], alpha=0.8, label="2-Phase+TTA+Opt")
b2 = ax.bar(x + width/2, [s["ens_acc"] for s in subgroup_stats], width,
            color=PALETTE["orange"], alpha=0.8, label="Ens+TTA+Opt")
for bar, a in zip(b1, [s["tp_acc"] for s in subgroup_stats]):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.005, f"{a:.3f}", ha="center", fontsize=8)
for bar, a in zip(b2, [s["ens_acc"] for s in subgroup_stats]):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.005, f"{a:.3f}", ha="center", fontsize=8)
ax.set_xticks(x); ax.set_xticklabels(names, fontsize=9)
ax.set_ylabel("Accuracy"); ax.set_title("Accuracy by Grade", fontweight="bold")
ax.legend(fontsize=9); ax.set_ylim([0, 1.12])

# Row 1: Errors + scatter + agreement
ax = axes[1, 0]
b1 = ax.bar(x - width/2, [s["tp_errors"] for s in subgroup_stats], width,
            color=PALETTE["twophase"], alpha=0.8, label="2-Phase+TTA+Opt")
b2 = ax.bar(x + width/2, [s["ens_errors"] for s in subgroup_stats], width,
            color=PALETTE["orange"], alpha=0.8, label="Ens+TTA+Opt")
for bar, e in zip(b1, [s["tp_errors"] for s in subgroup_stats]):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.1, str(e), ha="center", fontsize=10, fontweight="bold")
for bar, e in zip(b2, [s["ens_errors"] for s in subgroup_stats]):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.1, str(e), ha="center", fontsize=10, fontweight="bold")
ax.set_xticks(x); ax.set_xticklabels(names, fontsize=9)
ax.set_ylabel("Errors"); ax.set_title("Errors by Grade", fontweight="bold"); ax.legend(fontsize=9)

ax = axes[1, 1]
for g in grade_list:
    mask = df_analysis["dr_grade"] == g
    ax.scatter(df_analysis.loc[mask, "prob_tp"], df_analysis.loc[mask, "prob_ens"],
              c=grade_colors[g], s=25, alpha=0.6, label=grade_names[g])
ax.plot([0, 1], [0, 1], "k--", alpha=0.3); ax.set_xlabel("2-Phase+TTA P(DR)"); ax.set_ylabel("Ensemble+TTA P(DR)")
ax.set_title("Probability Correlation", fontweight="bold"); ax.legend(fontsize=8)

ax = axes[1, 2]
agree_correct = ((y_pred_tp_best == y_pred_ens_best) & (y_pred_ens_best == y_true)).sum()
agree_wrong   = ((y_pred_tp_best == y_pred_ens_best) & (y_pred_ens_best != y_true)).sum()
tp_only       = ((y_pred_tp_best == y_true) & (y_pred_ens_best != y_true)).sum()
ens_only      = ((y_pred_tp_best != y_true) & (y_pred_ens_best == y_true)).sum()
both_diff     = ((y_pred_tp_best != y_true) & (y_pred_ens_best != y_true) & (y_pred_tp_best != y_pred_ens_best)).sum()
agree_data = {
    "Both Correct": agree_correct, "Both Wrong (agree)": agree_wrong,
    "Only 2-Phase Correct": tp_only, "Only Ensemble Correct": ens_only,
    "Both Wrong (disagree)": both_diff,
}
colors_ag = ["#27ae60", "#c0392b", PALETTE["twophase"], PALETTE["orange"], "#7f8c8d"]
bars = ax.barh(list(agree_data.keys()), list(agree_data.values()), color=colors_ag, edgecolor="white")
for bar, v in zip(bars, agree_data.values()):
    ax.text(bar.get_width()+0.5, bar.get_y()+bar.get_height()/2,
            f"{v} ({v/len(y_true)*100:.1f}%)", va="center", fontsize=9)
ax.set_xlabel("Count"); ax.set_title("Model Agreement", fontweight="bold"); ax.invert_yaxis()

plt.tight_layout()
plt.savefig(FIG_DIR / "subgroup_analysis.png", dpi=300, bbox_inches="tight")
plt.show()
print(f"Saved → {FIG_DIR / 'subgroup_analysis.png'}")
```
